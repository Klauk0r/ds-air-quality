```python
import statsmodels.api as sm

import statsmodels.formula.api as smf

from sklearn.metrics import r2_score, mean_squared_error

data = sm.datasets.get_rdataset("dietox", "geepack").data

X = data[['Time', 'Start']]
grp = data[['Pig']]
y = data.Weight

md = sm.MixedLM(data.Weight,data[['Time', 'Start']],data.Pig)

mdf = md.fit()

weight_pred = md.predict(mdf.fe_params, exog=data[['Time', 'Start']])

print(mdf.summary())

print('R^2:', round(r2_score(data.Weight, weight_pred),2))
print('RMSE:', round(mean_squared_error(data.Weight, weight_pred, squared=False),2))
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()

model = linreg.fit(data[['Time', 'Start']],data.Weight)

weight_pred = linreg.predict(data[['Time', 'Start']])

#print(model.summary())

print('R^2:', round(r2_score(data.Weight, weight_pred),2))
print('RMSE:', round(mean_squared_error(data.Weight, weight_pred, squared=False),2))
```
